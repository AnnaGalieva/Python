**График

```python
from sympy import*
from sympy.plotting import plot
init_printing()
x = Symbol('x')
f = 18*x**3 + 5*x**2 + 10*x - 30
plot(f)
print(f)
```

**Вершина

```python
def parabola(p, q):
    x = -p / 2
    y = x * x + p * x + q
    return x, y
    
p=10
q=-30
x, y = parabola(p, q)
print('Вершина: ({:.3f} ,  {:.3f})'.format(x, y)) 
```
